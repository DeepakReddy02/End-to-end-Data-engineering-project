Silver Data

```python
Silver_Data = spark.read.parquet("/Volumes/workspace/rainfall_data/silver_layer/RainFall_data")
Silver_Header = spark.read.csv("/Volumes/workspace/rainfall_data/silver_layer/RainFall_header")
display(Silver_Data)
display(Silver_Header)
```

```python
display(Silver_Data.describe())
display(Silver_Header.describe())
```

```python
Silver_Data_Count = Silver_Data.count() - Silver_Data.dropDuplicates().count()
print(f"Number of duplicate rows in Data: {Silver_Data_Count}")
```

```python
Silver_Header_Count = Silver_Header.count() - Silver_Header.dropDuplicates().count()
print(f"Number of duplicate rows in Data: {Silver_Header}")
```

```python
#check for distinct values
for column in Silver_Data.columns:
    print(f"{column}: {Silver_Data.select(column).distinct().count()} distinct values")
```

```python
from pyspark.sql.functions import col, sum
#Null value check
display(Silver_Data.select(*(sum(col(c).isNull().cast("int")).alias(c) for c in Silver_Data.columns)))
```
